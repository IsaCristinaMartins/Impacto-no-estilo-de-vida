1. Ao visualizar a base você percebeu que seria melhor alterar o nome de
algumas colunas. Mude o ‘ID’ para ‘Identificador’, corrija o nome da

2. Qual é a média, a moda e a mediana de horas de sono para cada uma
das profissões? [‘mean’, np.median, pd.Series.mod]

```python
#Organizando o dataframe para extrair média, moda e mediana.
df_profissoes = df_saude_sono_novo[['Profissão', 'Duração do sono']]
df_profissoes
```

```python
#Contando os valores individuais

df_profissao = df_saude_sono_novo['Profissão'].value_counts()
print('Essa é a quantidade de profissoinais por profissão')
df_profissao
```

```python
#média
print('Essa é a média de sono por profissão (h)')
df_profissoes_media = df_saude_sono_novo.groupby('Profissão')['Duração do sono'].mean()
print(df_profissoes_media)

df_media_geral = df_saude_sono_novo['Duração do sono'].mean()
print('Essa é a média geral de duração do sono:', df_media_geral)
```

```python
#mediana
print('Essa é a soma e apresentação crescente de horas de sono por profissão')
df_profissoes_soma = df_saude_sono_novo.groupby('Profissão')['Duração do sono'].sum()

print(df_profissoes_soma.sort_values())

print('Essa é a mediana de sono por profissão')
df_profissoes_soma.median()
```

```python
#moda
print('Essa é a moda de sono por profissão')
df_moda = df_profissoes_soma.mode()

print(df_moda)
```

8 - Quem tem maior pressão sanguínea média, homens ou mulheres?
(Considere a média)

```python
df_saude_sono_novo[['Pressão sistólica', 'Pressão diastólica']] = df_saude_sono_novo['Pressão Sanguínea'].str.split('/', expand=True).astype(int)

PAM = (df_saude_sono_novo['Pressão sistólica'] + 2 *df_saude_sono_novo['Pressão diastólica'])/3
df_saude_sono_novo['Pressão Arterial Média'] = PAM
media_genero = df_saude_sono_novo.groupby('Gênero')['Pressão Arterial Média'].mean().idxmax()

print("O gênero com maior pressão é: ", media_genero)
```

10 -  Pessoas com frequências cardíacas acima de 70 dão mais passos que
pessoas com frequência cardíaca menor ou igual a 70? (Use a média)

```python

media_mais_70 = df_saude_sono_novo[df_saude_sono_novo['Frequência cardíaca'] > 70]['Passos diários'].mean()
media_menos_70 = df_saude_sono_novo[df_saude_sono_novo['Frequência cardíaca'] <= 70]['Passos diários'].mean()

# Comparação das médias
if media_mais_70 > media_menos_70:
    print("pessoas com frequência cardíaca acima de 70 dão mais passos.")
elif media_mais_70 < media_menos_70:
    print("Não, pessoas com frequência cardíaca igual ou menor a 70 dão mais passos.")
else:
    print("Não, pessoas com frequência cardíaca acima de 70 e Pessoas com frequência cardíaca igual ou menor a 70 dão o mesmo número de passos.")
```
